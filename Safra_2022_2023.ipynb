```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Safra 2022/2023
safras = {
    'Cultura': ['Soja', 'Milho', 'Cana-de-açúcar', 'Café', 'Algodão', 'Arroz', 'Feijão', 'Trigo'],
    'Produção (milhões de toneladas)': [154.5, 125.7, 610.1, 10.1, 2.98, 11, 3.3, 9.5],
    }
```

```python
# df
df = pd.DataFrame(safras)
df
```

```python
# Comentário referenciando as fontes oficiais
# Fontes dos dados:
# - Companhia Nacional de Abastecimento (CONAB). (2023). Acompanhamento da Safra Brasileira de Grãos. Disponível em: https://www.conab.gov.br
# - Instituto Brasileiro de Geografia e Estatística (IBGE). (2023). Produção Agrícola Municipal (PAM). Disponível em: https://www.ibge.gov.br
# - Ministério da Agricultura, Pecuária e Abastecimento (MAPA). (2023). Anuário Estatístico da Agricultura. Disponível em: https://www.gov.br/agricultura/pt-br
```

```python
# Produção culturas - Safra 2022/2023
figsize = (12, 1.2 * len(df['Cultura'].unique()))
cmap = sns.color_palette("Greens", as_cmap=True)
norm = plt.Normalize(df['Produção (milhões de toneladas)'].min(), df['Produção (milhões de toneladas)'].max())

plt.figure(figsize=figsize)
ax = sns.barplot(df,
                 x='Produção (milhões de toneladas)',
                 y='Cultura',
                 palette=cmap(norm(df['Produção (milhões de toneladas)'])))
sns.despine(top=True,
            right=True,
            bottom=True,
            left=True)

for i in ax.containers:
    ax.bar_label(i,
                 fontsize=14,
                 label_type='edge')

ax.set(xlabel=None)
ax.set(xticks=[])
ax.set(ylabel=None)
plt.yticks(fontsize=16)
```
